# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost

    return m

```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

```python
p=[5,10,3,12,5,50,6]
table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

```python
# optimal parenthesization

def matrix_chain_order(arr):
   n = len(arr)
   dp = [[0] * n for _ in range(n)]
   split = [[0] * n for _ in range(n)]
   for length in range(2, n):
       for i in range(1, n - length + 1):
           j = i + length - 1
           dp[i][j] = sys.maxsize
           for k in range(i, j):
               cost = dp[i][k] + dp[k+1][j] + arr[i-1] * arr[k] * arr[j]
               if cost < dp[i][j]:
                   dp[i][j] = cost
                   split[i][j] = k
   return dp, split
def print_optimal_parens(split, i, j):
   if i == j:
       return chr(65 + i - 1)
   return "(" + print_optimal_parens(split, i, split[i][j]) + \
          print_optimal_parens(split, split[i][j] + 1, j) + ")"

dp, split = matrix_chain_order(p)
print("Optimal Parenthesization:", print_optimal_parens(split, 1, len(p)-1))
```

```python
# number of multiplications saved

def count_normal_multiplications(p):
    first=p[0]
    last=p[2]
    res=p[0]*p[1]*p[2]
    for i in range(3,len(p)):
        res+=first*last*p[i]
        last=p[i]
    return res

normal_mult=count_normal_multiplications(p)
optimal_mult= table[1][len(p)-1]

print("Multiplications saved:",normal_mult-optimal_mult)
```
